```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
def generate_svg(model_name, prompt, device):
    try:
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        outputs = model.generate(**inputs, max_new_tokens=256)
        svg_code = tokenizer.decode(outputs[0], skip_special_tokens=True)
        return svg_code
    except Exception as e:
        print(f"Error with {model_name}: {e}")
        return None
```

**Модель PowerLM-3B**

[huggingface](https://huggingface.co/ibm/PowerLM-3b)





PowerLM-3B — это современная небольшая языковая модель размером 3 миллиарда параметров, обученная с помощью шедуллера Power.

```python
model_name = "ibm/PowerLM-3b"
```

```python
prompt = "Generate a valid SVG code snippet that draws a circle with red. The SVG should have a width and height of 200px, and no title. The generated SVG code should not include XML header, just the raw SVG tags. The opening tag is: <svg width='200' height='200'>"
```

```python
svg_code = generate_svg(model_name, prompt, device)
if svg_code:
  print(f"Generated SVG:\n{svg_code}\n")

  try:
    with open(f"generated_{model_name.replace('/', '_')}.svg", "w") as f:
      f.write(svg_code)
      print(f"SVG saved to generated_{model_name.replace('/', '_')}.svg")
  except Exception as e:
      print(f"Error saving SVG: {e}")
else:
    print(f"Failed to generate SVG for {model_name}")
```
